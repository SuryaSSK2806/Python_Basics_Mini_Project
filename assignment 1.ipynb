```python
def add_student(students, student_id, name, marks):
    student = {
        "id": student_id,
        "name": name,
        "marks": marks
    }
    students.append(student)
    print("Student details added successfully!\n")


def calculate_avg(marks):
    return sum(marks) / len(marks)


def find_student(students, student_id):
    for student in students:  # ✅ fixed loop syntax
        if student['id'] == student_id:
            return student
    return None


def display_all_students(students):
    if not students:
        print("No student records found.\n")
    else:
        for student in students:
            avg = calculate_avg(student["marks"])  # ✅ fixed function name
            print(f"ID: {student['id']}, Name: {student['name']}, Average: {avg:.2f}")
        print()  # ✅ moved inside block


# ------------------- Main Program -------------------
students = [
    {"id": 1, "name": "Alice", "marks": [80, 85, 88]},
    {"id": 2, "name": "Bob", "marks": [60, 70, 77]}
]

while True:
    print("1. Add Student")
    print("2. View All Students")
    print("3. Search Student by ID")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        student_id = int(input("Enter ID: "))
        name = input("Enter Name: ")
        marks = list(map(int, input("Enter Marks (3 subjects): ").split()))
        add_student(students, student_id, name, marks)

    elif choice == "2":
        display_all_students(students)

    elif choice == "3":
        student_id = int(input("Enter Student ID to search: "))
        student = find_student(students, student_id)
        if student:
            avg = calculate_avg(student["marks"])
            print(f"ID: {student['id']}, Name: {student['name']}, Average: {avg:.2f}\n")
        else:
            print("Student not found.\n")

    elif choice == "4":
        print("Exiting program. Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.\n")

```
